```python
import numpy as np
import pandas as pd
```

```python
df_raw = pd.read_csv("https://github.com/zhouhe018/damg6105/releases/download/raw_data/US_youtube_trending_data_raw.csv")
df_raw.sample()
```

```python
cate_raw = pd.read_json("https://github.com/zhouhe018/damg6105/releases/download/raw_data/US_category_id.json")
cates = cate_raw['items']
```

```python
id2cate = {}
for cate in cates:
    ids = int(cate["id"])
    snippet = cate["snippet"]
    if ids not in id2cate:
        id2cate[ids] = snippet["title"]
id2cate 
```

```python
# add category column
df_raw["category"] = df_raw.categoryId.apply(lambda x: id2cate[x] if x in id2cate else "Unknown")

# delete columns that are not useful this time
df_us = df_raw.drop(columns = ['categoryId','thumbnail_link','comments_disabled','ratings_disabled'])

# rename columns to keep the consistency
df_us.columns = ['video_id', 'video_title', 'publish_on', 'channel_id', 'channel_title','trend_on', 'tags', 'view_count','likes', 'dislikes', 'comment_count', 'description','category' ]

# convert the timestamp into readable date format
df_us['publish_on'] = pd.to_datetime(df_us['publish_on']).dt.date
df_us['trend_on'] = pd.to_datetime(df_us['trend_on']).dt.date
```
